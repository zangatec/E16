```
# Incluye aquí todos módulos, librerías y paquetes que requieras.
import numpy as np
import pandas as pd

from sklearn.decomposition import TruncatedSVD
```

#**Ejercicio-1.** 

```
#Se extraen columnas de interés de archivo "rating_final.csv" (previamente cargado en sección de archivos):

total_rat_rest = pd.read_csv("/content/rating_final.csv", header=0, sep=",")   

print(total_rat_rest.shape)
total_rat_rest.head(5)
```

```
#A su vez, se extraen columnas de interés de archivo "geoplaces2.csv" (también previamente cargado en sección de archivos):

rest_info = pd.read_csv('/content/geoplaces2.csv', header=0, encoding='latin-1', sep=",")

rest_name = rest_info[['placeID', 'name']] 
print(rest_name.shape)
rest_name.head(5)

#rest_info.head(5)
```

```
#Matriz de utilidad con valores de columna "rating":

rating_rest = total_rat_rest[['userID', 'placeID', 'rating']]  # seleccionemos las columnas con las que trabajaremos.

rating_merge = pd.merge(rating_rest, rest_name, on= 'placeID')

Ut_rating = rating_merge.pivot_table(values='rating', index='userID', columns='name', fill_value=0)

print("Tamaño de matriz de utilidad con valores de Rating general: ", Ut_rating.shape)
Ut_rating.head(10)
```

#**Ejercicio-2.**

```
#SVD para rating:

X_rating = Ut_rating.T
print("Tamaño de matriz transpuesta de matriz de utilidad, según rating:", X_rating.shape)

SVD_rating = TruncatedSVD(n_components = 137, random_state = 1)
SVD_rating.fit(X_rating)

print("Cantidad de valores singulares obtenidos: ", ((SVD_rating.singular_values_).sum()).round(3))
```

#**Ejercicio-3.**

```
#Correlación de Pearson considerando valores de rating:

matrix_rating = SVD_rating.fit_transform(X_rating)
corrRating = np.corrcoef(matrix_rating,)
print("Tamaño de matriz de correlación de Pearson con valores de rating: ", corrRating.shape)
```

```
#Recomendaciones con rating geenral para usuario que evaluó bien "Subway":
like_rating = "Subway"

names_Rat = Ut_rating.columns
rating_list = list(names_Rat)
id_like_Rating = rating_list.index(like_rating)
print("ID de restaurante Subway: ", id_like_Rating)

corr_recom_rating = corrRating[id_like_Rating]
print("Tamaño de matriz de correlación de Subway con otros restaurantes: ", corr_recom_rating.shape)
print("Matriz de correlación de Subway con otros restaurantes: ", corr_recom_rating)
```

```
#Recomendaciones para usuarios que les gusta restaurante "Subway" de acuerdo a rating:
listRat_orig = list(names_Rat[(corr_recom_rating > .23) & (corr_recom_rating < 1)])
print("Lista de las 10 mejores recomendaciones para alguien que evaluó bien Subway, considerando rating:")
listRat_orig
```

```
#Se observa que al ver la correlación de la matriz de Subway de otros restaurantes con su propio ID es igual a 1 (como se esperaba):
corr_recom_rating[id_like_Rating]
```

#**Ejercicio-4.**

```
#Disminuyendo la dimensión de la factorización de SVD, considerando rating,
#para poder recomendar a usuarios que gustan de "Subway":

num_SVD_rat = 10

SVD_rat_red = TruncatedSVD(n_components=num_SVD_rat, random_state=1)

matrix_rat_red = SVD_rat_red.fit_transform(X_rating)
print("Tamaño de matriz reducida, considerando rating: ", matrix_rat_red.shape)

corrRat_red = np.corrcoef(matrix_rat_red, )
print("Tamaño de matriz de correlación reducida, considerando rating: ", corrRat_red.shape)

corr_recom_rat_red = corrRat_red[id_like_Rating]
print("Tamaño de matriz de correlación reducida recomendada para los que gustan de Subway (considerando rating): ",corr_recom_rat_red.shape)
print("Matriz de correlación reducida recomendada para los que gustan de Subway (considerando rating):", corr_recom_rat_red)
```

```
#Las recomendaciones considerando reducir el número de componentes, buscando que coincidan 9 de 10 recomendaciones cuando se consideran todos los valores (según el rating) es:
listRat_red = list(names_Rat[(corr_recom_rat_red > .9992) & (corr_recom_rat_red < 1)])
print("Tamaño de lista de recomendaciones de matriz reducida, según rating: ", len(listRat_red))
print("Lista de recomendaciones de matriz reducida, según rating:")
listRat_red
```

```
#Datos comunes entre lista con SVD original (n_components = 137) y SVD reducido, con food_rating:
common_servlist = set(listRat_orig).intersection(listRat_red)
print("Lista de restaurantes en común con SVD tamaño original y tamaño reducido, según rating: ")
print(common_servlist)
print("Cantidad de elementos comunes entre ambas listas, según rating: ", len(common_servlist))
```

```
#Cantidad de valores singulares encontrado y cantidad de información que contienen, considerando los valores de rating:
print('Cantidad de información simplificada con los primeros %d vectores singulares, considerando valores de rating:' % num_SVD_rat)
print('%.1f%%' %  (100 * (1- (SVD_rating.singular_values_[0:num_SVD_rat]).sum() / (SVD_rating.singular_values_).sum())))
```
